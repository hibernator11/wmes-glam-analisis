# Análisis de los eventos realizados en Wikimedia España

```python
import pandas as pd

df = pd.read_csv('GLAM-WMES-2025.csv')
```

```python
df
```

```python
df.dtypes
```

#### Vamos a convertir el campo fecha a tipo fecha para poder analizar los años

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format='mixed', errors='coerce')
```

```python
df
```

```python
df['Ano'] = df['Fecha'].apply(lambda x: x.strftime('%Y'))
```

```python
df.groupby(['Ano'])['Ano'].count()
```

#### Con pandas podemos crear una gráfica que muestre los eventos agrupados por año

```python
import matplotlib.pyplot as plt

df.groupby(['Ano'])['Ano'].count().plot.bar()
plt.xlabel('Año')
plt.ylabel('Núm. Eventos');
```
